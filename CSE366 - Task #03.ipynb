 **Digital-analog Conversion Techniques**

```python
#2017331064


import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def init_sinWave( t ,A,f, phase=0 ):
  x= A * np.sin(2.0*np.pi*f*t+  phase )
#  print(x)
  return x  
```

## **ASK** *(ampliltude shift keying)*



*   A = amplitude = 5
*   Carrier freq = 40
*   Channel Response freq = 3


```python

a=5
carrier_freq=40
chan_freq=3

t=np.arange(0,1,1/1000)
x=init_sinWave(t,3,carrier_freq, a/2)
u=(a/2)*signal.square(2 * np.pi * chan_freq * t)+(a/2)
v=x*u


fig, axs = plt.subplots(3, figsize=(25, 10))
axs[0].plot(x, linewidth=2)
axs[0].set(ylabel='Amplitude', title='Carrier')

axs[1].plot(u, linewidth=2)
axs[1].set(ylabel='Amplitude', title='Main')

axs[2].plot(t, v, linewidth=4)
axs[2].set(xlabel='Time',ylabel='Amplitude', title='ampliltude shift keying')
axs[2].grid()

plt.show()
```

## **PSK** *(phase shift keying)*

*   A = amplitude = 4
*   Carrier freq = 40
*   Channel Response freq = 3

```python

carrier_freq=50
chan_freq=25
a=4

t=np.arange(0,1,1/1000)

x= init_sinWave(t, a, carrier_freq )
# x=a*np.sin(2*np.pi*carrier_freq*t)
u=signal.square(2 * np.pi * chan_freq * t)
v=x*u


fig, axs = plt.subplots(3, figsize=(20, 15))
axs[0].plot(t,x, linewidth=3)
axs[0].set(ylabel='Amplitude', title='Carrier')
axs[0].grid()

axs[1].plot(u, linewidth=3)
axs[1].set(ylabel='Amplitude', title='Main')
axs[1].grid()

axs[2].plot(t, v, linewidth=3)
axs[2].set(xlabel='Time',ylabel='Amplitude', title='PSK')
axs[2].grid()
plt.show()
```

## **FSK** *(Frequency shift keying)*

```python
F1=35
a=10
F=10

t=np.arange(0,2,1/1500)
x=init_sinWave(t,a, F1)


u=signal.square(2*np.pi*F*t)+1
fsk=F1+F1*u/2
x2=a*np.sin(2*np.pi*fsk*t)

fig, axs = plt.subplots(3, figsize=(30, 10))
axs[0].plot(t,x, linewidth=3)
axs[0].grid()

axs[1].plot(u, linewidth=3)
axs[1].grid()

axs[2].plot(t,x2, linewidth=3)
axs[2].set(xlabel='Time',ylabel='Amplitude', title='FSK')
axs[2].grid()
plt.show()
```
