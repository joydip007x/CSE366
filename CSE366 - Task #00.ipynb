# Signal-01: Signals and Plotting the signal.

```python
import numpy as np
import matplotlib.pyplot as plt


#2017331064


t,Fs=np.linspace(-1, 3, 100, retstep=True)
#print(t,Fs)
```

 **Function**:  Creating Sinusoidal

---



$$x(t) = A\sin(2\pi f t + phase)$$


```python
def init_sinWave( t ,A,f, phase=0 ):
  x= A * np.sin(2.0*np.pi*f*t+  phase )
#  print(x)
  return x  
```

 **Function**:  Plotting Sinusoidal & Magnitude

---



```python

x = init_sinWave(t,3,1)


fig,axis = plt.subplots(nrows=2, ncols=2, figsize=(15,15))
fig.suptitle('Plotting the Sinusoidal Wave', fontsize=12)

axis[0,0].plot(t,x)
axis[0,0].set_title('Continuous Signal Plot',fontsize=13)

axis[0,1].scatter(t,x)
axis[0,1].set_title('Scattered Plot of Signal',fontsize=13)

axis[1,0].stem(t,x)
axis[1,0].set_title('Stem Plot of Signal ',fontsize=13)


axis[1,1].magnitude_spectrum(x,Fs,color='C1')
axis[1,1].set_title('Magnitude of the  Signal Plot',fontsize=13)


plt.show();
```

# Signal-02: Basic Properties of a Signal.

```python
t=np.arange(-1.0,5.0,0.00001)

x1= init_sinWave(t,2,0.1)
x2= init_sinWave(t,2,1)
x3= init_sinWave(t,2,10)

#Composite SIgnal
y = x1 + x2 +x3 ;
```

A **Composite** Sinusoidal Singal 

---



```python
fig,axis = plt.subplots(nrows=2, ncols=2, figsize=(15,15))

axis[0,0].plot(t,x1)
axis[0,0].set_title('Signal # 1', fontsize = 15)

axis[0,1].plot(t,x2)
axis[0,1].set_title('Signal # 2', fontsize = 15)

axis[1,0].plot(t,x3)
axis[1,0].set_title('Signal # 3', fontsize = 15)

axis[1,1].plot(t,y)
axis[1,1].set_title( 'Signal 1+2+3 ',fontsize= 16)

plt.show();
```

```python
fs=10 
step= 1/fs 

n=np.arange(-1,4,step)

y= init_sinWave(n,1,3) + init_sinWave(n,1,6) + init_sinWave(n,1,9)
plt.title('Sampling Freq=10', fontsize =17)
plt.stem(n,y)
plt.show()

fs=25 
step= 1/fs 

n=np.arange(-1,4,step)

y= init_sinWave(n,1,3) + init_sinWave(n,1,6) + init_sinWave(n,1,9)
plt.title('Sampling Freq=25', fontsize =17)
plt.stem(n,y)
plt.show()
```
