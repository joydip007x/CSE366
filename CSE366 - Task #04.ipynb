## **PM (Phase Moudlation)**

```python
#2017331064

import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import signal
```

```python

t=np.arange(0,2,0.001)
fc=10
fm=1
phase_deviation=np.pi/2

mt=0.5*(1+signal.square(2*np.pi*fm*t))
ct=np.sin(2*np.pi*fc*t)
pm=np.sin(2*np.pi*fc*t+phase_deviation*mt)


fig, axs = plt.subplots(3, figsize=(25, 10)) 
plt.subplot(3,1,1)
plt.title('Phase Modulation')
plt.plot(t,ct,linewidth=3)
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')
plt.grid('on')

plt.subplot(3,1,2)
plt.plot(t,mt,linewidth=3)
plt.ylabel('Amplitude')
plt.xlabel('Message signal')
plt.grid('on')

plt.subplot(3,1,3)
plt.plot(t,pm,linewidth=3)
plt.ylabel('Amplitude')
plt.xlabel('PM signal')
plt.grid('on')
```

## **FM (Frequency Moudlation)**

```python
modulator_frequency = 4.0
carrier_frequency = 40.0
modulation_index = 1.0

time = np.arange(44100.0) / 44100.0
modulator = np.sin(2.0 * np.pi * modulator_frequency * time) * modulation_index
carrier = np.sin(2.0 * np.pi * carrier_frequency * time)
product = np.zeros_like(modulator)

for i, t in enumerate(time):
    product[i] = np.sin(2. * np.pi * (carrier_frequency * t + modulator[i]))


fig, axs = plt.subplots(3, figsize=(25, 10)) 

plt.subplot(3, 1, 1)
plt.title('Frequency Modulation')
plt.plot(modulator,linewidth=3)
plt.ylabel('Amplitude')
plt.xlabel('Modulator signal')
plt.subplot(3, 1, 2)
plt.plot(carrier,linewidth=3)
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')
plt.subplot(3, 1, 3)
plt.plot(product,linewidth=3)
plt.ylabel('Amplitude')
plt.xlabel('Output signal')
plt.show()
```
